### Load Data

```python
import pandas as pd
import numpy as np

url='https://raw.githubusercontent.com/galawana/DSA_Python/master/Google_Play_Datasets/googleplaystore.csv'
df = pd.read_csv(url, encoding='utf-8')

df.head(2)
```

### Data Cleansing

```python
min_rating = df['Rating'].min()
max_rating = df['Rating'].max()
print("rating from ",min_rating," to ",max_rating,", it's should be between 0 .0 to 5.0, so there is some missplace on index ",df[df['Rating'] == 19].index)
print("\ndata index 10472 is shifting to left from column Category onwards, so we should shift 1 to right from Category")

df.iloc[10472,1:] = df.iloc[10472,1:].shift(1)
```

## Main Case

### 1. How many rows and columns reflected on dataset file googleplaystore.csv?

```python
print("dataset file googleplaystore.csv shape is ",df.shape)
print("kolom : ", len(df.columns),", baris : ",len(df))
```

#### further cleaning data

```python
print("\nRemove Duplicates Apps")
df.drop_duplicates(subset='App', inplace=True)
print("\nRemove + and commas from installs and change to Int")
df['Installs'] = df['Installs'].apply(lambda x: x.replace('+', '') if '+' in str(x) else x)
df['Installs'] = df['Installs'].apply(lambda x: x.replace(',', '') if ',' in str(x) else x)
df['Installs'] = df['Installs'].astype(int)

print("\nChange rating to float")
df['Rating'] = df['Rating'].astype(float)
print("\nChange Reviews to Int")
df['Reviews'] = df['Reviews'].astype(int)

df.info()
```

### 2. How many unique app categories that cover in this dataset? 

```python
category = df[['Category']]
print("Unique Category : ",len(category.groupby(['Category'])))

res = pd.get_dummies(df['Genres'].str.split(';').apply(pd.Series).stack()).sum(level=0)
print("Unique Genre : ",len(res.columns))
```

### 3. What are top 3 apps that have highest number of users installed/download among all categories? 

```python
apps_install = df.sort_values('Installs',ascending=False)
apps_install.loc[:,['App','Category','Installs']].head(3)
```

### 4. What are top 3 apps that have highest number of users installed/download among the following categories

```python
list_of_category = ["BUSINESS", "COMMUNICATION", "GAME", "DATING", "MUSIC"]
data_category = df[df['Category'].isin(list_of_category)]
data_install = data_category.groupby(['Category'])
data_install = data_install.apply(lambda x: x.sort_values(["Installs"],ascending=False))
data_install = data_install[['App','Category','Installs']].reset_index(drop=True)
data_install.groupby(['Category']).head(3)
```

### 5. What are top 3 apps that have highest rating among the following categories:

```python
data_rating = data_category.groupby(['Category'])
data_rating = data_rating.apply(lambda x: x.sort_values(["Rating"],ascending=False))
data_rating = data_rating[['App','Category','Rating']].reset_index(drop=True)
data_rating.groupby(['Category']).head(3)
```

### 6. If you use two criteria (installs and rating), what are top 3 apps for the following categories? 

```python
data_both = data_category.groupby(['Category'])
data_both = data_both.apply(lambda x: x.sort_values(["Installs","Rating"],ascending=False))
data_both = data_both[['App','Category','Installs','Rating']].reset_index(drop=True)
data_both.groupby(['Category']).head(3)
```

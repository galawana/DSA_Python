### What Central of  measurement to class

```python
df_class = df.iloc[:,4:]
df_class.groupby('class',as_index=True).mean().transpose()
```

* Class 0 memiliki mean values International_plan, voice_mail_plan, number_vmail_messages yang lebih tinggi dibandingkan class 1
* sedangkan Class 0 memiliki nmber_customer_service_calls, total_day_minutes, total_day_charge yang lebih rendah dibandingkan dengan class 1
* untuk measurement lain relative memiliki mean yang sama

### What Measure of Dispersion from group class

```python
df_class = df.iloc[:,4:]
df_class.groupby('class',as_index=True).std().transpose()
```

* Hampir semua measure memiliki dispersion yang relative mirip untuk class 0 dan 1
* hanya internation_plan, voice_mail_plan, total_day_minutes, dan number_customer_service_calls yang memiliki dispersion yang membedakan antara class 0 dan 1
* dengan kolom total_eve_minutes, total_day_minutes, total_night_minutes memiliki dispersion terbesar

### Correlation all measures to Class

```python
df_class = df.iloc[:,4:]
abs(df_class.corr(method='kendall').loc['class'].sort_values(ascending=False))
```

* korelasi tertinggi adalah international_plan dengan value 0.259, dan semua measurement features memiliki korelasi yang rendah

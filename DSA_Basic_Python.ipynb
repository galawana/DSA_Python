## Use Case: High Rating Movies Criteria
Hints:
* Lets set imdb_score benchmark as High-Rating movie with imdb_score > 7.5
* Create new variables score_group (High if imdb_score > 7.5 and Low if imdb <= 7.5)
* Analyze at least 5 variables from which you can start figure out criteria of high rated movie
* Analyse 5 more variables to strengthen criteria of high rated movies

```python
def rating(row):
    if row['imdb_score'] > 7.5 : return 'High'
    return 'Low'

imdb['score_group'] = imdb.apply(lambda row : rating(row), axis=1)
imdb['score_group2'] = np.where(imdb['imdb_score'] > 7.5,'High','Low')
```

```python
high_rated = imdb[imdb.score_group == 'High']
```
